```python
def quality(total_bill,tip):
    if tip/total_bill >0.25:
        return "generous"
    else:
        return "other"
```

```python
df['quality']=np.vectorize(quality)(df['total_bill'],df['tip'])
#same as what is done in above cell but easier --> transforming a function which isn't aware of numpy array to one which is aware
```
