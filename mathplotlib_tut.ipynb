```python
fig =plt.figure(figsize=(10,10),facecolor='pink')
#large axes
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,9)
axes1.set_ylim(0,9000)
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.set_title('power of 4')
#samll axes
axes2 = fig.add_axes([0.2,0.6,0.19,0.29])
#keep para less than 1 n greater than 0 to make it inside the previous axes
axes2.plot(x,y)
axes2.set_xlim(0,9)
axes2.set_ylim(0,10)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('twice')
plt.show()
```

```python
fig = plt.figure(figsize=(1,1),dpi=100)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
#dpi is to clear bluriness
fig.savefig('newfigure.png',bbox_inches='tight')
#bbox_inches is used to save axes in fig too
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label="x vs x")
ax.plot(x,y, label="twice")
ax.legend(loc='upper right')
#loc = location where we want legends to be located
#ax.legend(loc=(1.1,0.5))--> manual putting--> is outside box
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='red',lw=2,linestyle='-.',marker='o',markersize=15,markerfacecolor='blue',markeredgewidth=8,markeredgecolor='green')
ax.plot(x,x+9,color='#e6c017',linewidth=2,ls=':',marker='s',ms=8)
#we can pass rgb hex code for colors
#lw --> line width & ls -->linestyle & ms--> markersize
#possible linestyple options '--','-','-.',':','steps' 
#marker is to depict our data points
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lines =ax.plot(x,x,color='red',lw=5)
lines[0].set_dashes([3,2,1,1,3,2])
#since line is a 2d object list--> we are passing a list that
#corresponds with our custom dash sequence--> how many solid points we want then how many blanks 
#then how many solids then how many blanks.. n so on
```
